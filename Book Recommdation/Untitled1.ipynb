## Popularity based Recommendation System

```python
# SO here we have to find top 100 books by its rating. 
```

```python
books_with_ratings = ratings.merge(books, on='ISBN')
books_with_ratings
```

```python
num_ratings = books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.rename(columns={'Book-Rating' : 'Num-Rating'}, inplace=True)
num_ratings
```

```python
avg_ratings = books_with_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={'Book-Rating' : 'Avg-Rating'}, inplace=True)
avg_ratings
```

```python
final_ratings = avg_ratings.merge(num_ratings, on='Book-Title')
final_ratings
```

```python
top_books = final_ratings[final_ratings['Num-Rating'] > 250]
top_books
```

```python
top_books = top_books.sort_values('Avg-Rating', ascending=False)
top_books
```

```python
top_100_books = top_books.head(100)
top_100_books
```

```python
top_100_books.duplicated().sum()
```

```python
top_100_books = top_100_books.merge(books, on='Book-Title')
top_100_books = top_100_books.drop_duplicates('Book-Title')
top_100_books
```

```python
top_100_books.drop(columns=['ISBN', 'Year-Of-Publication', 'Publisher', 'Image-URL-S', 'Image-URL-L'], inplace=True)
top_100_books
```

```python
top_100_books['Book-Title'][0]
```

```python
pickle.dump(top_100_books, open('popular.pkl', 'wb'))
```

## Collaborative Filtering Based recommadation System

```python
# Here We have to make Recommendation system according to the Users Ratings.
# So will choose Those users which have rated more than 250 books.
# And will Choose Those books which have recieved at least 50 ratings. 
```

```python
books_with_ratings
```

```python
all_users = books_with_ratings.groupby('User-ID').count()['Book-Rating'] > 250
top_users = all_users[all_users].index
top_users
```

```python
rated_by_top_users = books_with_ratings[books_with_ratings['User-ID'].isin(top_users)]
rated_by_top_users
```

```python
all_books = rated_by_top_users.groupby('Book-Title').count()['Book-Rating'] > 50
top_books = all_books[all_books].index
top_books
```

```python
final_ratings = rated_by_top_users[rated_by_top_users['Book-Title'].isin(top_books)]
final_ratings
```

```python
df = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
df
```

```python
df.fillna(0, inplace=True)
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(df)
similarity_scores.shape
```

```python
similarity_scores[0]
```

```python
def recommend(book_name):
    index = np.where(df.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:5]
    data = []
    for i in similar_item:
        suggestion = []
        temp_df = books[books['Book-Title'] == df.index[i[0]]]
        suggestion.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        suggestion.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        suggestion.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(suggestion)
        
    return data
```
